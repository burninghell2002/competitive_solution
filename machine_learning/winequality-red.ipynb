Importing the nexessary things

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

Importing data set

```python
df=pd.read_csv("/content/drive/MyDrive/datasets/winequality-red.csv",delimiter=';')
df
```

Check for nan - no na values found

```python
df.describe()
```

Visualizing data

```python
graph=plt.scatter(df.quality,df.iloc[:,:1] )
```

making x and y

```python
x=df.iloc[:,:11]
y=df.iloc[:,11]
```

Getting numpy arrays

```python
x=x.to_numpy()
x=preprocessing.maxabs_scale(x)
```

```python
y=y.to_numpy()
y=preprocessing.maxabs_scale(y)
y.shape
```

```python
y=y.reshape(-1,1)
y.shape
```

train test split

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

Training the model

```python
model=LinearRegression().fit(x_train,y_train)
```

Checking accuracy

```python
score=model.score(x_test,y_test)
score
```
